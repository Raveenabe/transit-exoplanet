```python
import matplotlib.pyplot as plt

# Sample data
x = range(1, 101)
y = [i ** 2 for i in x]

# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot entire graph on the first subplot
ax1.plot(x, y)
ax1.set_title('Entire Graph')

# Plot zoomed-in portion on the second subplot
zoom_start = 20
zoom_end = 40
ax2.plot(x[zoom_start:zoom_end], y[zoom_start:zoom_end])
ax2.set_title('Zoomed-In View')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = range(1, 101)
y = [i ** 2 for i in x]

# Plot the entire graph
plt.plot(x, y, label='Entire Graph')

# Set limits for the zoomed-in region
zoom_start = 20
zoom_end = 40
plt.axis([zoom_start, zoom_end, min(y[zoom_start:zoom_end]), max(y[zoom_start:zoom_end])])

# Plot the zoomed-in region
plt.plot(x[zoom_start:zoom_end], y[zoom_start:zoom_end], label='Zoomed-In View', color='red')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Zoomed-In View of Part of the Graph')
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Sample data
x = range(1, 101)
y = [i ** 2 for i in x]

# Create the main plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Mark the point to zoom in on
zoom_point_x = 50
zoom_point_y = zoom_point_x ** 2
ax.scatter(zoom_point_x, zoom_point_y, color='red', label='Zoom Point')

# Zoomed-in view
zoom_width = 20
zoom_height = 1000
zoom_area = (zoom_point_x - zoom_width/2, zoom_point_x + zoom_width/2,
             zoom_point_y - zoom_height/2, zoom_point_y + zoom_height/2)

# Create inset axes
ax_inset = inset_axes(ax, width="30%", height="30%", loc='upper right')
ax_inset.plot(x, y)
ax_inset.set_xlim(zoom_area[0], zoom_area[1])
ax_inset.set_ylim(zoom_area[2], zoom_area[3])

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Zoomed-In View with Marked Point')
ax.legend()

plt.show()
```
